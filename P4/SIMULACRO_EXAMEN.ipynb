### 📝 PREGUNTA 1: Curvas Isoparamétricas y Tangentes

Dada la superficie paramétrica $S(u,v) = \left(\begin{matrix} v \cos(u) \\ v \sin(u) \\ v^2 \\ 1 \end{matrix}\right)$ definida para $u \in [0, 2\pi]$ y $v \in [0, 2]$. Consideramos la curva isoparamétrica $C_{u_0}(v)$ que se obtiene al fijar el parámetro $u_0 = \pi/2$.

Calcula las coordenadas homogéneas del vector tangente a dicha curva isoparamétrica en el punto donde $v = 2$.

**Opciones:**
- a) `[0, 2, 2, 0]`
- b) `[-2, 0, 4, 0]`
- c) `[0, 1, 4, 0]`
- d) `[1, 0, 4, 0]`

```python
# Espacio para tus cálculos en SageMath
u, v = var('u v')
```

<details>
<summary><b>💡 Solución y Razonamiento (Haz clic para ver)</b></summary>
<b>Respuesta correcta: C</b><br>
Si fijamos $u_0 = \pi/2$, sustituimos en $S$: $\cos(\pi/2)=0$ y $\sin(\pi/2)=1$.<br>
La curva queda: $C_{\pi/2}(v) = (0, v, v^2, 1)^T$.<br>
Para hallar el vector tangente, derivamos respecto a $v$:<br>
$dC/dv = (0, 1, 2v, 0)^T$. <i>(Recuerda que la coordenada homogénea de un vector director es 0, no 1).</i><br>
Evaluando en $v = 2$: $(0, 1, 4, 0)^T$.

### 📝 PREGUNTA 2: Vector Normal a la Superficie

Sea la superficie paramétrica (paraboloide hiperbólico) definida por la expresión $S(u,v) = \left(\begin{matrix} u \\ v \\ u^2 - v^2 \\ 1 \end{matrix}\right)$. 

Haciendo uso de la función `producto_vectorial(U,V)` definida en las prácticas, calcula el vector normal (sin normalizar) a la superficie en el punto de parámetros $u = 1$, $v = 1$. ¿Cuáles son sus coordenadas homogéneas?

**Opciones:**
- a) `[2, 2, -1, 0]`
- b) `[-2, 2, 1, 0]`
- c) `[-1, 1, 2, 0]`
- d) `[1, -2, 2, 0]`

```python
# Espacio para tus cálculos en SageMath
```

<details>
<summary><b>💡 Solución y Razonamiento (Haz clic para ver)</b></summary>
<b>Respuesta correcta: B</b><br>
El vector normal se calcula como el producto vectorial de los vectores tangentes parciales $S_u \times S_v$.<br>
Derivada respecto a $u$: $S_u = (1, 0, 2u, 0)^T$. Evaluado en $u=1 \rightarrow S_u = (1, 0, 2, 0)^T$<br>
Derivada respecto a $v$: $S_v = (0, 1, -2v, 0)^T$. Evaluado en $v=1 \rightarrow S_v = (0, 1, -2, 0)^T$<br>
Haciendo el producto vectorial (determinante $i, j, k$):<br>
x = (0)(-2) - (2)(1) = -2<br>
y = - [(1)(-2) - (2)(0)] = 2<br>
z = (1)(1) - (0)(0) = 1<br>
Por tanto, el vector es `[-2, 2, 1, 0]`.
</details>

### 📝 PREGUNTA 3: Superficies de Revolución

Considera una curva generatriz situada en el plano XZ definida paramétricamente como $C(u) = \left(\begin{matrix} u \\ 0 \\ u^2 \\ 1 \end{matrix}\right)$. Generamos una **superficie de revolución** rotando esta curva alrededor del eje Z un ángulo $v$. 

Si evaluamos la superficie resultante en $u = 2$ y $v = \pi$, ¿qué punto obtenemos?

**Opciones:**
- a) `[-2, 0, 4, 1]`
- b) `[2, 0, -4, 1]`
- c) `[0, -2, 4, 1]`
- d) `[2, 2, 4, 1]`

```python
# Espacio para tus cálculos en SageMath
```

<details>
<summary><b>💡 Solución y Razonamiento (Haz clic para ver)</b></summary>
<b>Respuesta correcta: A</b><br>
Para crear la superficie de revolución, se multiplica la curva por la matriz de rotación en Z:<br>
$S(u,v) = M_{Rot\_Z}(v) \cdot C(u)$<br>
$x = u \cdot \cos(v)$<br>
$y = u \cdot \sin(v)$<br>
$z = u^2$<br>
Evaluando en $u=2$ y $v=\pi$:<br>
$x = 2 \cdot \cos(\pi) = 2 \cdot (-1) = -2$<br>
$y = 2 \cdot \sin(\pi) = 2 \cdot (0) = 0$<br>
$z = 2^2 = 4$<br>
El punto en coordenadas homogéneas es `[-2, 0, 4, 1]`.
</details>
